```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
json_df = pd.read_json("./data_final.json")
json_df.describe()
```

```python
countries = json_df["Country"].unique()

fig = plt.figure(1, figsize=(100, 170), dpi=80)

position = 1

X_predict = np.arange(1995, 2020).reshape(-1,1)

for country in countries:
    
    y = json_df[["GDI"]][json_df["Country"]==country].to_numpy()
    X = json_df[["Year"]][json_df["Country"]==country].to_numpy()
    # print(country, X, y)

    # Create a linear model
    model = LinearRegression()

    # Fit (Train) our model to the data
    model.fit(X,y)

    predictions = model.predict(X_predict)

    ax = fig.add_subplot(17, 10, position)

    position = position + 1

    plt.scatter(X, y, c='blue')
    plt.plot(X_predict, predictions, c="red")
    plt.title(country)

plt.show()



    
```
